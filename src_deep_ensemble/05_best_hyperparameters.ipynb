```python
import json
import pandas as pd
trial= 0

hp_res = pd.DataFrame()

for trial in range(0,20):

    with open(f'./results/DE_HP_4000/result_trial_{trial}.json') as f:

        data = json.load(f) 
        row = pd.DataFrame(data['params'], index=[trial])
        row['accuracy'] = data['accuracy']
        row['loss'] = data['value']

        hp_res = pd.concat([hp_res, row])
```

```python
hp_res.sort_values('accuracy', ascending=False, inplace=True)
```

```python
hp_res.to_csv('./results/DE_HP_4000/results_4000_top_5.csv')
```
